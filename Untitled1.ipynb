```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
```

```python
#data preprocessing 
stopwords=stopwords.words('english')
new_stop_words=['movie','one','shall','could','would','flim','might']
stopwords.extend(new_stop_words)
stopwords.remove('not')
stopwords=set(stopwords)
```

```python
def remov_stopwords(text):
    tokens = word_tokenize(text)
    tokens = [token.strip() for token in tokens]
    filtered_tokens = [token for token in tokens if token not in stopwords]
    filtered_txt = ' '.join(filtered_tokens)    
    return filtered_txt
```

```python
def remove_special_char(text):
    text=re.sub(r'<.*?>','',text)
    text= re.sub(r'\[[^!\@#$%&]*\]',' ', text)
    text=re.sub(r'[^a-zA-z0-9\s]','',text)
    text= re.sub('https?://\S+|www\.\S+', ' ', text)
    #text= re.sub("[,\.!?:()']``",' ', text)
    return text
```

```python
def expand_words(content):
    content=re.sub(r"won\'t","would not",content)
    content=re.sub(r"\'ll","will",content)
    return content
```

```python
def data_preprocess(content):
    content=remove_special_char(content)
    content=remove_special_char(content)
    #content=remov_stopwords(content)
    content=expand_words(content)
    return content
```

```python
#df['review']=df['review'].apply(remov_stopwords)
```

```python
df['review']=df['review'].apply(data_preprocess)
```

```python
from wordcloud import WordCloud
text_series = df['review']

text_data = text_series.str.cat(sep=' ')

# Create WordCloud instance with custom parameters
wordcloud = WordCloud(background_color='white', max_words=200, width=800, height=400)

# Generate the word cloud
wordcloud.generate(text_data)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
